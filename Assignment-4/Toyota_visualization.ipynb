```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Toyota.csv")
data.head()
```

# Scatter plot

```python
plt.scatter(data['Age_08_04'],data['Price'])
plt.xlabel('Age_08_04')
plt.ylabel('Price')
plt.show()
```

#  Histogram

```python
data.hist(figsize = (10,20))
```

# Bar plot of fuel

```python
counts = [979, 120, 12]
Fuel_Type = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(Fuel_Type))
plt.bar(index, counts, color=['red', 'blue', 'cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, Fuel_Type, rotation = 90)
plt.show()
```

# Regression plot

```python
sns.set(style="darkgrid")
sns.regplot(x=data['Age_08_04'],y=data['Price'])
```

# Distplot

```python
sns.distplot(data['Age_08_04'])
```

# Count Plot

```python
sns.countplot(x="Fuel_Type", data=data)
```

```python
sns.countplot(x="Fuel_Type", data=data, hue = "Automatic")
```

# Box Plot

```python
sns.boxplot(y=data["Price"])
```

```python
sns.boxplot(x=data['Fuel_Type'], y=data["Price"])
```

# Pair plot

```python
sns.pairplot(data, kind="countplot")
plt.show()
```
